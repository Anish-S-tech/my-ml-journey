```python
# confusion matrix for binary classification (2 classes only)

import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
actual    = np.array(
  ['Dog','Dog','Dog','Not Dog','Dog','Not Dog','Dog','Dog','Not Dog','Not Dog'])
predicted = np.array(
  ['Dog','Not Dog','Dog','Not Dog','Dog','Dog','Dog','Dog','Not Dog','Not Dog'])
```

```python
# Precision and recall
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import precision_score, recall_score

encoder = LabelEncoder()

actual = encoder.fit_transform(actual)
predicted = encoder.fit_transform(predicted)

precision1 = precision_score(actual,predicted)
recall1 = recall_score(actual,predicted)

print(f"Precision: {precision1}")
print(f"Recall: {recall1}")
```

```python
confusion_mat = confusion_matrix(actual,predicted)
```

```python
sns.heatmap(confusion_mat, fmt='d', cmap="Blues", annot=True)
```

```python
print(classification_report(actual,predicted))
```

```python
# Confusion matrix for multiple classes - (3 or more classes)
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
y_true = ['Cat'] * 10 + ['Dog'] * 12 + ['Horse'] * 10
y_pred = ['Cat'] * 8 + ['Dog'] + ['Horse'] + ['Cat'] * 2 + ['Dog'] * 10 + ['Horse'] * 8 + ['Dog'] * 2
classes = ['Cat', 'Dog', 'Horse']
```

```python
print(y_true)
print(y_pred)
print(classes)
```

```python
confusion = confusion_matrix(y_true,y_pred)
```

```python
# one way - Using seaborn's heatmap

sns.heatmap(confusion,fmt='d', cmap="Blues", annot = True)
```

```python
# another way - Using matplotlib + sklearn (ConfusionMatrixDisplay)

confusion = confusion_matrix(y_true, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=confusion)
disp.plot(cmap="Blues")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import precision_score, recall_score

encoder = LabelEncoder()

y_true = encoder.fit_transform(y_true)
y_pred = encoder.fit_transform(y_pred)

precision2 = precision_score(y_true,y_pred,average='macro')
recall2 = recall_score(y_true,y_pred, average='micro')

print(f"Precision: {precision2}")
print(f"Recall: {recall2}")
```

```python
# F1 score

from sklearn.metrics import f1_score

f1_none = f1_score(y_true,y_pred,average=None)
f1_micro = f1_score(y_true,y_pred,average="micro")
f1_macro = f1_score(y_true,y_pred,average="macro")
f1_weighted = f1_score(y_true,y_pred,average="weighted")

print(f"f1 score per class: {f1_none}")
print(f"Micro-average f1 score: {f1_micro}")
print(f"Macro-average f1 score: {f1_macro}")
print(f"Weighted-average f1 score: {f1_weighted}")
```
