```python
# This file explains about the implementation of adaboost algorithm from scratch using iris dataset
```

```python
# Step 1: Importing librairies

import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris,make_classification
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Step 2: Defining the Adaboost class

class AdaBoost:
  def __init__(self,n_estimators=50):
    self.n_estimators = n_estimators
    self.alphas = []
    self.models = []

# Step 3: Training the adaboost model

  def fit(self, X,y):
    n_samples, n_features = X.shape
    w = np.ones(n_samples)/n_samples

    for _ in range(self.n_estimators):
      model = DecisionTreeClassifier(max_depth=1)
      model.fit(X,y,sample_weight=w)

      predictions = model.predict(X)

      err = np.sum(w* (predictions!=y))/np.sum(w)

      alpha = 0.5 *np.log((1-err)/(err+1e-10))

      self.models.append(model)
      self.alphas.append(alpha)

      w*=np.exp(-alpha*y*predictions)
      w/=np.sum(w)

# Step 4: Defining the predict method

  def predict(self, X):
      strong_preds = np.zeros(X.shape[0])

      for model, alpha in zip(self.models, self.alphas):
          predictions = model.predict(X)
          strong_preds += alpha * predictions

      return np.sign(strong_preds).astype(int)

```

```python
# Step 5: Model evaluation and real world usage

X, y = make_classification(n_samples=1000, n_features=20, n_classes=2, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

adaboost = AdaBoost(n_estimators=50)
adaboost.fit(X_train, y_train)

predictions = adaboost.predict(X_test)

classification = classification_report(y_test,predictions)
confusion = confusion_matrix(y_test,predictions)

print(confusion)
print(classification)
```
