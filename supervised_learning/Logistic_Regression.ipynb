```python
# Binomial Linear Regression implementation - (contains only two possible values for dependent variables)
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X, y = load_breast_cancer(return_X_y=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42,test_size=0.3)
```

```python
model = LogisticRegression(max_iter=1000,random_state=0)
model.fit(X_train,y_train)
```

```python
acc = accuracy_score(y_test,model.predict(X_test))*100
print(f"Logistic regression model accuracy: {acc:.2f}%")
```

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
```

```python
digits = load_digits()
```

```python
X = digits.data
y = digits.target
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.3)
```

```python
reg = LogisticRegression(max_iter=10000, random_state=0)
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)

print(f"Logistic Regression model accuracy: {accuracy_score(y_test, y_pred) * 100:.2f}%")
```

```python
# Cost function in logistic regression - log loss function (cross-entropy loss) implementation
```

```python
import numpy as np
```

```python
def sigmoid(z):
  return 1/(1+np.exp(-z))
```

```python
def cost_function(y,y_pred):
  m = len(y)
  cost = -1/m * np.sum(y * np.log(y_pred) + (1-y) * np.log(1-y_pred))
  return cost
```

```python
# Example dataset

X = np.array([0.2, 0.4, 0.6])
y = np.array([0, 1, 1])
```

```python
# Random weights

theta = 0.5
z = X * theta
y_pred = sigmoid(z)
```

```python
print("Predicted Probabilities:", y_pred)
print("Cost Function Value:", cost_function(y, y_pred))
```
