```python
# Create a custom dataframe

player_list = [['M.S.Dhoni', 36, 75, 5428000],
               ['A.B.D Villers', 38, 74, 3428000],
               ['V.Kohli', 31, 70, 8428000],
               ['S.Smith', 34, 80, 4428000],
               ['C.Gayle', 40, 100, 4528000],
               ['J.Root', 33, 72, 7028000],
               ['K.Peterson', 42, 85, 2528000]]

players = pd.DataFrame(player_list, columns=['Name','Age','Weight','Runs'])  # Giving column names using 'columns' parameter
print(players)
```

```python
data = players[players['Age']>35].iloc[0:3]   # Slicing based on conditions
print(data)
```

```python
display(players.loc[(players['Name'] == 'A.B.D Villers') & (players['Age'] > 35) & (players['Weight'] > 70)])  # '&' used for combining multiple conditions
```

```python
import numpy as np

filtered = np.where((players['Name'] == 'A.B.D Villers') & (players['Age'] > 35) & (players['Weight'] > 70))  # filtering based on multiple conditions using numpy
print(filtered)  # returns the datatype of result and which index it's matching
```

```python
data1 = {'Name': ['Jai', 'Princi', 'Gaurav', 'Anuj'],
         'Age': [27, 24, 22, 32],
         'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
         'Qualification': ['Msc', 'MA', 'MCA', 'Phd']}

data2 = {'Name': ['Abhi', 'Ayushi', 'Dhiraj', 'Hitesh'],
         'Age': [17, 14, 12, 52],
         'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
         'Qualification': ['Btech', 'B.A', 'Bcom', 'B.hons']}

df = pd.DataFrame(data1, index=[0, 1, 2, 3])

df1 = pd.DataFrame(data2, index=[4, 5, 6, 7])
```

```python
import pandas as pd

data1 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'key1': ['K0', 'K1', 'K0', 'K1'],
         'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32],}

data2 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'key1': ['K0', 'K0', 'K0', 'K0'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']}

df = pd.DataFrame(data1)

df1 = pd.DataFrame(data2)


print(df, "\n\n", df1)
```

```python
# Normalization is the process of transforming values to a standard scale without distorting differences in the ranges of values

# normalized_value = (value - min_value) / (max_value - min_value)

# all values lies between 0 to 1 after normalization

import matplotlib.pyplot as plt

df = pd.DataFrame([
    [180000, 110, 18.9, 1400],
    [360000, 905, 23.4, 1800],
    [230000, 230, 14.0, 1300],
    [60000, 450, 13.5, 1500]
], columns=['Col A', 'Col B', 'Col C', 'Col D'])

display(df)

df.plot(kind='box')
plt.show()
```

```python
# normalization techniques

# 1. maximum absolute scaling (also called as Decimal scaling)

# formula = x/(x.max()).abs()  [x / maximum absolute value of x]

max_scaled = df.copy()

for column in max_scaled.columns:
  max_scaled[column] = df[column] / df[column].abs().max()

display(max_scaled)
max_scaled.plot(kind='bar')
plt.show()
```

```python
# 2. min-max normalization (also called as feature scaling)

# formula = x - max(x) / max(x) - min(x)

min_max_scaled = df.copy()

for column in min_max_scaled.columns:
  min_max_scaled[column] = (df[column] - df[column].min()) / (df[column].max() - df[column].min())

display(min_max_scaled)
min_max_scaled.plot(kind='bar')
plt.show()
```

```python
# 3. z-score method (also called as standardization)

# Converts data into values with mean as 0 and standard deviation as 1

# formula: z = (x - mean(x)) / std(x)

z_scaled = df.copy()

for column in z_scaled.columns:
  z_scaled[column] = (z_scaled[column] - z_scaled[column].mean()) / z_scaled[column].std()

display(z_scaled)
z_scaled.plot(kind = 'bar')
plt.show()
```
