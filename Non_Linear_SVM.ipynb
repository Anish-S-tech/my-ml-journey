```python
# 1. Below is the implementation of non linear svm in circular decision boundary
```

```python
# Step 1: Importing the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
# Step 2: Creating and splitting the dataset

X, y = make_circles(n_samples=500,factor=0.5,noise=0.05, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42,test_size=0.3)
```

```python
# Step 3: Create and Train Non linear SVM model

svm = SVC(kernel='rbf', C=1, gamma=5)
svm.fit(X_train, y_train)
```

```python
# Step 4: Make predictions and evaluate the model

y_pred = svm.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
print(f"Accuracy: {accuracy :.2f}")
```

```python
# Step 5: Visualizing the decision boundary

def plot_decision_boundary(X,y,model):
  X_min, X_max = X[:,0].min()-1, X[:,0].max()+1
  y_min, y_max = X[:,1].min()-1, X[:,1].max()+1

  xx, yy = np.meshgrid(np.arange(X_min,X_max,0.01), np.arange(y_min,y_max,0.01))

  Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
  Z = Z.reshape(xx.shape)

  plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.Paired)
  plt.scatter(X[:,0], X[:,1], c=y, edgecolor='k', cmap=plt.cm.Paired)

  plt.title("Non linear SVM with RBF kernel")
  plt.show()


plot_decision_boundary(X,y,svm)
```

```python
# Step 1: Importing libraries

import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# Step 2: Creating and splitting the dataset

X, y = make_moons(n_samples=500, noise=0.1, random_state=42)
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
# Step 3: Creating and training the SVM with polynomial kernel

svm_poly = SVC(C=1, kernel='poly', degree=3, coef0=1)   # degree and coef0 controls the curve
svm_poly.fit(X_train, y_train)
```

```python
# Step 4: Making predictions and evaluating the model

y_pred = svm_poly.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"Accuracy score: {accuracy:.2f}")
```

```python
# Step 5: Visualizing the decision boundary

def plot_decision_boundary(X,y,model):
  X_max, X_min = X[:,0].max()+1, X[:,0].min()-1
  y_max, y_min = X[:,1].max()+1, X[:,1].min()-1

  xx, yy = np.meshgrid(np.arange(X_min, X_max, 0.01), np.arange(y_min, y_max, 0.01))

  Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
  Z = Z.reshape(xx.shape)

  plt.contourf(xx,yy, Z, alpha=0.8, cmap=plt.cm.Paired)
  plt.scatter(X[:,0], X[:,1], c=y, edgecolor='k',cmap=plt.cm.Paired)

  plt.title("Non linear SVM with polynomial kernel")
  plt.show()

plot_decision_boundary(X,y,svm_poly)
```
